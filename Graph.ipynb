```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pyspark.ml.clustering import KMeans


pddf_pred = pd.read_csv("pandas_csv.csv") 
pddf_pred.loc[:,'score'] *= 2500
```

```python
pddf_pred.head()
```

```python

# Languages
languages = ["JavaScript", "Java", "PHP", "Python", "C#", "C++", "Ruby", "CSS",
      "Objective-C", "Perl", "Scala", "Haskell", "MATLAB", "Clojure", "Groovy"]

#Colors
colors = ['red','green','blue', 'teal', 'olive', 'purple', 'maroon','navy', 'aqua', 'yellow', 'gray','black']
```

```python
threedee = plt.figure(figsize=(30,20)).gca(projection='3d')
threedee.scatter(pddf_pred.language, pddf_pred.prediction, c=pddf_pred.score)
threedee.set_xlabel('language')
threedee.set_ylabel('prediction')
threedee.set_zlabel('score')
threedee.set_xticklabels(languages)


plt.show()
```

```python
ax = plt.figure(figsize=(15,10)).gca(projection='3d')
x =pddf_pred.language
y =pddf_pred.prediction
z =pddf_pred.score

X,Y = np.meshgrid(np.arange(10), np.arange(10))
Z = np.sin(X) + np.sin(Y)

ax.set_xlabel('language')
ax.set_ylabel('prediction')
ax.set_zlabel('score')
ax.set_xticklabels(languages)


colors =plt.cm.magma( (X-X.min())/float((X-X.min()).max()) )
ax.plot_trisurf(x,y,z ,  cmap="magma")

plt.show()
```
